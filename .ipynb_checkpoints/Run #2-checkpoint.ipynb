```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup #Importing BS4 Beta
import requests
```

```python
site = 'https://mcphd-tx.org/Covid-19-Case-Updates/'
page = requests.get(site)
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
print(soup.prettify())
```

```python
test = soup.findAll('tbody')
```

```python
test[0].contents[1].contents[1].contents[0]
```

It looks like even though displaying all the data requires a button press, the HTML may have all the data already? Since our last case displayed is number 1428, and the limit is 1000 cases at a time.

```python
for i, n in enumerate(test[0].contents[1]):
    print(i, n)
```
